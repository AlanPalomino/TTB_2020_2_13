```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sn
from sklearn import metrics
import matplotlib.pyplot as plt
```

<h2>Linear Analysis<h2>

<h2>Healthy group<h2>

```python
healthy_linear = pd.read_csv("Test/linear_healthy.csv",index_col=0)
healthy_linear.dropna(inplace=True)

healthy_linear = healthy_linear[["mean","variance","skewness","kurtosis"]]
healthy_linear_id = np.repeat(0,71)
healthy_linear
```

<h2>ill group<h2>

```python
sick_linear = pd.read_csv("Test/linear_sick.csv",index_col=0)
sick_linear.dropna(inplace=True)

sick_linear = sick_linear[["mean","variance","skewness","kurtosis"]]
sick_linear_id = np.repeat(1,195)
sick_linear
```

<h2>Scatter plots<h2>

```python
fig = plt.figure(figsize=(7,5))

a = plt.scatter(healthy_linear["mean"],healthy_linear["kurtosis"],color = "steelblue",alpha=0.5)
b = plt.scatter(sick_linear["mean"],sick_linear["kurtosis"],color = "red",alpha =0.5)

plt.title("Gráfico de dispersión (sesgo vs Curtosis)")
plt.ylabel("curtosis")
plt.xlabel("sesgo")
plt.legend((a,b),("Sano","ECV"))
plt.axis

fig.savefig("graf/s1/skwvskur(1).png")
```

<h2>Train data sets<h2>

```python
targets = healthy_linear_id.tolist() + sick_linear_id.tolist()
healthy_linear_train = healthy_linear.iloc[:,0:4]
sick_linear_train = sick_linear.iloc[:,0:4]
X=pd.concat([healthy_linear_train,sick_linear_train],ignore_index=True)
X
```

<h2>Split test and train groups<h2>

```python
#Split data to test and train groups
X_train,X_test,y_train,y_test = train_test_split(X,targets,test_size=0.20,random_state=1)
print(len(X_train))
print(len(y_train))
X.dropna()
X
```

<h2>Logistic regression<h2>

```python
from sklearn.preprocessing import StandardScaler

#Logistic Regression function
logistic_regression= LogisticRegression(max_iter=20000)
sc = StandardScaler()

#Training
X_train = sc.fit_transform(X_train)
logistic_regression.fit(X_train,y_train)
X_test = sc.fit_transform(X_test)

#Testing
y_pred=logistic_regression.predict(X_test)
y_test=np.array(y_test)

#Confussion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
```

<h2>Define train set and targets<h2>

```python
targets = healthy['cond_id'].tolist() + pathologys['cond_id'].tolist()
healthy_train = healthy.iloc[:,1:24]
pathologys_train = pathologys.iloc[:,0:22]
X=pd.concat([healthy_train,pathologys_train],ignore_index=True)
X
```

<h2>Split train data sets <h2>

```python
#Split data to test and train groups
X_train,X_test,y_train,y_test = train_test_split(X,targets,test_size=0.20,random_state=1)
print(len(X_train))
print(len(y_train))
X.dropna()
X
```

<h2>Logistic Regression<h2>

```python
from sklearn.preprocessing import StandardScaler

#Logistic Regression function
logistic_regression= LogisticRegression(max_iter=20000)
sc = StandardScaler()

#Training
X_train = sc.fit_transform(X_train)
logistic_regression.fit(X_train,y_train)
X_test = sc.fit_transform(X_test)

#Testing
y_pred=logistic_regression.predict(X_test)
y_test=np.array(y_test)

#Confussion matrix
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
```

<h2>Random forest<h2>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler


classifier = RandomForestClassifier()
sc = StandardScaler()
X_train = sc.fit_transform(X_train)

classifier.fit(X_train, y_train)

X_test = sc.fit_transform(X_test)
y_pred = classifier.predict(X_test)
print("Accuracy : ", accuracy_score(y_test, y_pred))

confusion_matrix = pd.crosstab(np.array(y_test), np.array(y_pred), rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

ns_probs = [0 for _ in range(len(y_test))]
lr_probs = classifier.predict_proba(X_test)
lr_probs = lr_probs[:,1]

ns_auc = roc_auc_score(y_test, ns_probs)
lr_auc = roc_auc_score(y_test, lr_probs)

print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Logistic: ROC AUC=%.3f' % (lr_auc))

ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)

fig = plt.figure(figsize=(10,6))

plt.plot(ns_fpr, ns_tpr, linestyle='--', label='Sin entrenamiento')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Extra Trees')

plt.title("Sanos vs ECV")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
fig.savefig('ROC_HealthyvsECV.png')
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc

fig = plt.figure(figsize=(10,6))
lr_probs = classifier.predict_proba(X_test)
lr_probs = lr_probs[:, 1]
yhat = classifier.predict(X_test)

lr_precision, lr_recall, _ = precision_recall_curve(y_test, lr_probs)
lr_f1, lr_auc = f1_score(y_test, yhat), auc(lr_recall, lr_precision)

print('Extra trees: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
y_test = str(y_test)
no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='Sin entrenamiento')
plt.plot(lr_recall, lr_precision, marker='.', label='Extra-trees')

plt.title('Sanos vs ECV')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
fig.savefig('r-c_sanovsECV.png')
```
